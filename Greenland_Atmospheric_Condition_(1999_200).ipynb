```python
# Import all the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io
from google.colab import drive
```

```python
loc = pd.read_csv('/content/drive/MyDrive/POWER_Regional_Monthly_1999_2000.csv')
new_loc = loc.drop(columns=['LAT', 'LON', 'PARAMETER', 'YEAR'])
new_loc.head()
```

```python
temp = new_loc.iloc[0:160]
temp.head()
```

```python
temp = new_loc.iloc[0:160]

# x and y data
months = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC", "ANN"]
values = temp

# Plot for the graph
plt.figure(figsize=(12, 8))
for i in range(len(values)):
    plt.plot(values.columns, values.iloc[i], color='#367F9A')

# Customizing the x and y axis to crop the graph exactly
plt.xlim(0, len(months) - 1)
plt.ylim(values.min().min(), values.max().max())

# Customizing the plot
plt.title("TEMPERATURE IN GREENLAND \n (1999‐2000)", fontsize=14, weight='bold')
plt.xlabel("Months & Annual", fontsize=12, weight='bold')
plt.ylabel("Temperature °C at 2m", fontsize=12, weight='bold')
plt.grid(alpha=0.5)

# Save the graph in the local drive
plt.savefig('temperature_in_greenland_1999_2000.png', dpi=300, bbox_inches='tight')

# Show plot
plt.tight_layout()
plt.show()
```

```python
spec_humid = new_loc.iloc[160:320]

# x and y data
months = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC", "ANN"]
values = spec_humid

# Plot for the graph
plt.figure(figsize=(12, 8))
for i in range(len(values)):
    plt.plot(values.columns, spec_humid.iloc[i], color='#9A367F')

# Customizing the x and y axis to crop the graph exactly
plt.xlim(0, len(months) - 1)
plt.ylim(values.min().min(), values.max().max())

# Customizing the plot
plt.title("SPECIFIC HUMIDITY IN GREENLAND \n (1999‐2000)", fontsize=14, weight='bold')
plt.xlabel("Months & Annual", fontsize=12, weight='bold')
plt.ylabel("Specific Humidity g/kg at 2m", fontsize=12, weight='bold')
plt.grid(alpha=0.5)

# Save the graph in the local drive
plt.savefig('Specific Humidity_in_greenland_1999_2000.png', dpi=300, bbox_inches='tight')

# Show plot
plt.tight_layout()
plt.show()
```
