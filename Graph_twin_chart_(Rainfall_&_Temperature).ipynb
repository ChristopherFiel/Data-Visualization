```python
# Import all the necessary packages
import matplotlib.pyplot as plt

# Datas for the Graph
categories = ['Jan', 'Feb', 'March', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
precipitation_values = [20, 15, 17, 20, 30, 70, 80, 65, 40, 30, 22, 20]
temperature_values = [15, 18, 25, 30, 33, 35, 37, 36, 28, 23, 15, 16]

# Create subplots with shared y-axis
fig, ax1 = plt.subplots()

# Bar graph for precipitation on the left side
ax1.bar(categories, precipitation_values, color='skyblue', label='Rainfall (mm)')

# Line graph for temperature on the right side
ax2 = ax1.twinx()
ax2.plot(categories, temperature_values, color='orange', marker='o', label='Temperature (Celsius)')

# Add labels and title
ax1.set_xlabel('Year 2020')
ax1.set_ylabel('Rainfall (mm)')
ax2.set_ylabel('Temperature (Celsius)')
plt.title('Monthly Temperature and Precipitation\nData in Synoptic Station 1')

# Show legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```
