```python
# Read the excel sheet
dlog = pd.read_excel('/content/drive/MyDrive/Copy of DATALOG.CSV.xlsx')
dlog.head()
```

```python
# Removes the Unit tab and rename the Temperature value with corresponding unit
new_dlog = dlog.drop(columns='Unit[C/F]')
new_dlog = new_dlog.rename(columns={'Temp' : 'Temp (°C)'})
new_dlog.tail(15)
```

```python
# Load the specified rows for plotting
road = new_dlog.iloc[793:798]

# Manually convert the time into pandas data frame
time = ['12:19', '12:24', '12:29', '12:34', '12:39']

# Create a twin chart for all the data
fig, ax1 = plt.subplots()

# Bar graph for Relative Humidity (RH)
ax1.bar(time, road['RH[%]'], color='green', alpha=0.6, label='RH[%]')

# Plot Temperature on the same axis as RH
ax1.plot(time, road['Temp (°C)'], color='royalblue', marker='o', linewidth=2, linestyle='dashed', label='Temp (°C)')

# Create a second y-axis for CO2 values
ax2 = ax1.twinx()
ax2.plot(time, road['CO2[ppm]'], color='red', marker='o', linewidth=2, linestyle='dashed', label='CO2[ppm]')

# Set axis labels
ax1.set_xlabel('Time')
ax1.set_ylabel('Relative Humidity (%) & Temperature (°C)')
ax2.set_ylabel('CO2 (ppm)', color='red')

# Combine legends from both axes and place them in the lower left
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='lower left')

plt.title('Air Quality at the RoadSide of RTU (2024-10-01)')

plt.show()
```

```python
ambient = new_dlog.iloc[798:803]
ambient_time = ['12:44', '12:49', '12:54', '12:59', '13:04']

# Create a twin chart for all the data
fig, ax1 = plt.subplots()

# Bar graph for Relative Humidity (RH)
ax1.bar(ambient_time, ambient['RH[%]'], color='green', alpha=0.6, label='RH[%]')

# Plot Temperature on the same axis as RH
ax1.plot(ambient_time, ambient['Temp (°C)'], color='royalblue', marker='o', linewidth=2, linestyle='dashed', label='Temp (°C)')

# Create a second y-axis for CO2 values
ax2 = ax1.twinx()
ax2.plot(ambient_time, ambient['CO2[ppm]'], color='red', marker='o', linewidth=2, linestyle='dashed', label='CO2[ppm]')

# Set axis labels
ax1.set_xlabel('Time')
ax1.set_ylabel('Relative Humidity (%) & Temperature (°C)')
ax2.set_ylabel('CO2 (ppm)', color='red')

# Combine legends from both axes and place them in the lower left
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='lower left')

plt.title('Air Quality at the Registrar of RTU (2024-10-01)')

plt.show()
```

```python
# assume `road` and `time` are already defined as before

fig = go.Figure()

# --- Relative Humidity as bar ---
fig.add_trace(go.Bar(
    x=time,
    y=road['RH[%]'],
    name='RH (%)',
    marker_color='cyan',
    opacity=0.7
))

# --- Temperature as line ---
fig.add_trace(go.Scatter(
    x=time,
    y=road['Temp (°C)'],
    name='Temp (°C)',
    mode='lines+markers',
    marker=dict(symbol='circle', size=8, color='magenta'),
    line=dict(dash='dash', width=4)
))

# --- CO2 on secondary y-axis ---
fig.add_trace(go.Scatter(
    x=time,
    y=road['CO2[ppm]'],
    name='CO₂ (ppm)',
    mode='lines+markers',
    marker=dict(symbol='square', size=8, color='orange'),
    line=dict(dash='dot', width=4),
    yaxis='y2'
))

# Layout tweaks for dark theme and centered, bold title
fig.update_layout(
    template='plotly_dark',
    title=dict(
        text='<b>Air Quality at the Roadside of RTU (October 10, 2024)</b>',
        x=0.5,
        xanchor='center',
        font=dict(size=18, color='white')
    ),
    xaxis=dict(
        title='Time',
        tickangle=-45,
        showgrid=True
    ),
    yaxis=dict(
        title='<b>Relative Humidity (%) & Temp (°C)<b>',
        showgrid=False,
        gridcolor='gray'
    ),
    yaxis2=dict(
        title='<b>CO₂ (ppm)<b>',
        overlaying='y',
        side='right',
        showgrid=True
    ),
    legend=dict(
        x=0.01, y=0.99,
        bgcolor='rgba(0,0,0,0)',
        bordercolor='white',
        borderwidth=1
    ),
    margin=dict(l=60, r=60, t=100, b=80),
    paper_bgcolor='#1e1e1e',
    plot_bgcolor='#1e1e1e'
)

fig.show()

# Exporting graphs in Plotly
```
